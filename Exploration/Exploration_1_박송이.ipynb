# 1-11. 프로젝트 (1) load_digits : 손글씨를 분류

### (1) 필요한 모듈 import하기

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

### (2) 데이터 준비

```python
digits = load_digits()
```

### (3) 데이터 이해하기

```python
digits_data = digits.data
digits_label = digits.target
print(digits.target_names)
print(digits.DESCR)
```

### (4) train, test 데이터 분리

```python
X_train,X_test,y_train,y_test = train_test_split(digits_data,
                                                 digits_label,
                                                 test_size =0.2,
                                                 random_state=7)
```

### (5) 다양한 모델로 학습시켜보기

```python
from sklearn.tree import DecisionTreeClassifier    # Decision Tree
from sklearn.ensemble import RandomForestClassifier #Random Forest
from sklearn import svm #SVM
from sklearn.linear_model import SGDClassifier  #SGD Classifier
from sklearn.linear_model import LogisticRegression #Logistic Regression

#정확도 평가
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
#Decision Tree
decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train,y_train)
y_pred = decision_tree.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
print(classification_report(y_test, y_pred))
```

```python
#Random Forest
random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train,y_train)
y_pred = random_forest.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
print(classification_report(y_test, y_pred))
```

```python
#SVM
svm_model = SGDClassifier()
svm_model.fit(X_train,y_train)
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
print(classification_report(y_test, y_pred))
```

```python
#SGD Classifier
sgd_model = SGDClassifier()
sgd_model.fit(X_train,y_train)
y_pred = sgd_model.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
print(classification_report(y_test, y_pred))
```

```python
#Logistic Regression
logistic_model = LogisticRegression()
logistic_model.fit(X_train,y_train)
y_pred = logistic_model.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
print(classification_report(y_test, y_pred))
```

### (6) 모델을 평가해 보기

Accuracy는 올바르게 예측된 데이터의 수를 전체 데이터의 수로 나눈 값

Accuray 지표로 모델을 평가하는게 맞다.
왜냐하면 이 경우에서는 true, false보다는 Label인 0~9를 얼마나 맞게 분류했는지가 중요하기 때문이다. 

Precision은 모델이 True로 예측한 데이터 중 실제로 True인 데이터의 수

Recall은 실제로 True인 데이터를 모델이 True라고 인식한 데이터의 수

F1 score는 precision 과 recall의 조화평균
